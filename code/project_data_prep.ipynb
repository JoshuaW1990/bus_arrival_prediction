#### Analysis for the old data (2014)

The data record the location of the vehicle at different time point in the route with position in detail. However, it does'nt provide the specific time point when the bus arrive the stop. 

I used to think the layover_during progress represents the time spot for arriving at a stop. But I'm wrong. The actual meaning of the LAYOVERD_DURING is the waiting state at the start and the terminal stop in a route. It is not very critical for our data.


- It is possible to be improved by the GTFS schedule data
